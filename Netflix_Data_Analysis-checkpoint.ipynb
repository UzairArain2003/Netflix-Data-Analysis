```python
#import the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load the data

df = pd.read_csv("Netflix Dataset.csv")
print(df.head(10))
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
# Step 1: Drop rows with missing values in critical columns (no assignment needed with inplace=True)
df.dropna(subset=['type', 'release_year', 'rating', 'country', 'duration', 'date_added'], inplace=True)

# Step 2: Fill missing values in 'director' and 'cast' (inplace applied directly)
df['director'].fillna("Unknown", inplace=True)
df['cast'].fillna("Not Listed", inplace=True)

# Step 3: Check for remaining missing values
print(df.isnull().sum())
```

```python
# Count how many Movies and TV Shows
type_counts = df['type'].value_counts()
print(type_counts)

# Plot the bar chart
plt.bar(type_counts.index, type_counts.values,color = ['orange','skyblue'])
plt.title('Number of Movies and TV Shows on Netflix')
plt.xlabel('Type')
plt.ylabel('Counts')
plt.tight_layout()
plt.savefig('Movies_VS_TvShows.png')
plt.show()
```

```python
# Count how many shows/movies have each rating
ratings_counts = df['rating'].value_counts().head(10) #top10 bcz pie chart isnt capable to handle all.
print(ratings_counts)
plt.figure(figsize=(12, 10))
plt.pie(ratings_counts, labels=ratings_counts.index, autopct='%1.1f%%', startangle=90, textprops={'fontsize': 10})

plt.title('Percentage of Content Ratings')

plt.tight_layout()
plt.savefig('content_ratings_pie.png')
plt.show()
```

```python
df_movie =df[df['type'] == 'Movie'].copy()
#df_movie['duration_int'] = df_movie['duration'].str.replace('min','').astype(int) 
df_movie = df_movie[df_movie['duration'].str.contains('min', na=False)]
df_movie['duration_int'] = df_movie['duration'].str.replace('min', '', regex=False).str.strip().astype(int)

plt.hist(df_movie['duration_int'],color = 'purple', edgecolor ='black',bins = 30)
plt.title('Distribution Of Movie Duration')
plt.xlabel('Duration in Minutes')
plt.ylabel('Number Of Movies')
#plt.grid(True)
plt.tight_layout()
plt.savefig('Movie_duration_histogram.png')
plt.show()
```

```python
release_counts = df['release_year'].value_counts().sort_index()
plt.scatter(release_counts.index,release_counts.values,color ='red' )
plt.title('Release Year VS Number Of Shows')
plt.xlabel('Realease Years')
plt.ylabel('Number Of Shows')
#plt.grid(True)
plt.tight_layout()
plt.savefig('release_year_scatter.png')
plt.show()
```

```python

top_countries = df['country'].value_counts().head(10)

sns.barplot(x=top_countries.values,y=top_countries.index,hue=top_countries.index,dodge=False,palette="muted",legend=False)

plt.title("Top 10 Countries with Most Netflix Content")
plt.xlabel("Number of Titles")
plt.ylabel("Country")
plt.savefig('Countries_content.jpg')
plt.show()
```

```python
content_by_years = df.groupby(['release_year','type']).size().unstack().fillna(0)
fig, ax =plt.subplots(1,2,figsize=(12, 5))
#first subplot For Movies
ax[0].plot(content_by_years.index,content_by_years['Movie'], color = 'orange')
ax[0].set_title('Movie Realeased Per Year')
ax[0].set_xlabel('Year')
ax[0].set_ylabel('Number Of Movies')

#second subplot for TV Shows
ax[1].plot(content_by_years.index,content_by_years['TV Show'], color = 'green')
ax[1].set_title('TV Show Realeased Per Year')
ax[1].set_xlabel('Year')
ax[1].set_ylabel('Number Of TV Shows')

fig.suptitle('Comparison Of Movies And TV Shows Released Over Years')
plt.savefig("movies_tvshows_comparison.jpg")
```

```python
sns.boxplot(x = df_movie['duration_int'], color="skyblue")
plt.title("Distribution of Movie Durations")
plt.xlabel("Minutes")
plt.show()
```

```python
# Probability of Movie vs TV Show
total = len(df)
prob_movie = len(df[df['type']=="Movie"]) / total
prob_show = len(df[df['type']=="TV Show"]) / total
print("P(Movie) =", prob_movie)
print("P(TV Show) =", prob_show)
print('If you randomly pick 1 title from Netflix dataset:\nThere‚Äôs a 69% chance it‚Äôs a Movie.\nThere‚Äôs a 31% chance it‚Äôs a TV Show.')


```

```python
prob_usa = len(df[df['country']=="United States"]) / total
print("P(Title from USA) =", prob_usa)
```

```python
numeric_df = df_movie[['release_year', 'duration_int']]
print(numeric_df.corr())

sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
